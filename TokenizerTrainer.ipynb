```python
import sentencepiece as spm
```

```python
#Train the tokenizer on text data
spm.SentencePieceTrainer.train(
    input='alllines.txt',       # your training text file
    model_prefix='mymodel',   # prefix for output files
    vocab_size=8000,          # number of tokens in vocab
    model_type='bpe',         # or 'unigram', 'word', 'char'
    character_coverage=1.0    # for non-English text, set lower
)
```

```python
#testing it
sp = spm.SentencePieceProcessor(model_file='mymodel.model')
```

```python
text = "Hello, how are you ?"
# Get token ids
ids = sp.encode(text, out_type=int)
print(ids)

# Get tokens as strings
tokens = sp.encode(text, out_type=str)
print(tokens)
```

```python
# Convert ids back to string
decoded_text = sp.decode(ids)
print(decoded_text)
```
